## get data

```python
data = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv')
data
```

## Use Matplotlib

```python
import matplotlib.pyplot as mat 
```

## Mayplotlib: Line graph

```python
data.plot()
mat.show()
```

## Matplotlib: Select

```python
data['station_paris'].plot()
mat.show()
```

## Matplotli: Scatter

```python
data.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
mat.show()      
```

## Matplotlib: Plot + Box

```python
data.plot.box()
mat.show()
```

## Matplotlib: Area

```python
axs = data.plot.area(figsize=(12, 4), subplots=True)
mat.show()
```

## REMEMBER
The .plot.* methods are applicable on both Series and DataFrames.

By default, each of the columns is plotted as a different element (line, boxplot,â€¦).

Any plot created by pandas is a Matplotlib object.
