```python
from matplotlib.pyplot import imread
import matplotlib.pyplot as plt

img = imread('Pic/rimuru tempest.jpg')
plt.imshow(img)
plt.show()

print("img Shape:",img.shape)

# 3 Arrray it mean RGB : Red Green Blue
img_red = img[:,:,0]
print("Red",img_red)

img_red = img[:,:,1]
print("Green",img_red)

img_red = img[:,:,2]
print("Blue",img_red)

# Or
print("\n Or...")
# Red
plt.imshow(img[:, :, 0], cmap='Reds')
plt.title('Red:')
plt.show()

# Green
plt.imshow(img[:, :, 1], cmap='Greens')
plt.title('Green:')
plt.show()

# Blue
plt.imshow(img[:, :, 2], cmap='Blues')
plt.title('Blue:')
plt.show()
```
